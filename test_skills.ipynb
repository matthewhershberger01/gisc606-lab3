```python
import os
import matplotlib.pyplot as plt
import seaborn as sns
import geopandas as gpd
import rioxarray as rxr 
# Plotting extent is used to plot raster & vector data together
from rasterio.plot import plotting_extent

import earthpy as et
import earthpy.plot as ep

# Prettier plotting with seaborn
sns.set(font_scale=1.5, style="white")
```

```python
et.data.get_data("colorado-flood")
os.chdir(os.path.join(et.io.HOME, 'earth-analytics', 'data'))
```

```python
import rioxarray as rxr
import matplotlib.pyplot as plt


raster_path = "colorado-flood/spatial/boulder-leehill-rd/pre-flood/lidar/pre_DTM_hill.tif"
raster = rxr.open_rasterio(raster_path)

plt.figure(figsize=(10, 5))
raster.plot(cmap="gray")
plt.title("Hillshade Overlay (ChatGPTs Version)")
plt.show()
```

```python
hillshade_path = "colorado-flood/spatial/boulder-leehill-rd/pre-flood/lidar/pre_DTM_hill.tif"
hillshade = rxr.open_rasterio(hillshade_path)

fig, ax = plt.subplots(figsize=(10, 5))
ep.plot_bands(hillshade, cmap="gray", ax=ax)
ax.set_title("Hillshade - Pre-flood")

ep.plot_bands(raster, cmap="viridis", alpha=0.5, ax=ax)
ax.set_title("DTM with Hillshade Overlay (ChatGPTs Version)")

plt.show()
```

```python
# Open the site boundary shapefile
site_bound_shp_path = "path_to_your_site_boundary_shapefile.shp"
site_bound_shp = gpd.read_file(site_bound_shp_path)

# Plot the hillshade layer and the DTM layer
fig, ax = plt.subplots(figsize=(10, 5))
ep.plot_bands(hillshade, cmap="gray", ax=ax)
ep.plot_bands(raster, cmap="viridis", alpha=0.5, ax=ax)

# Plot the site boundary layer
site_bound_shp.plot(ax=ax, facecolor="none", edgecolor="red")

# Set the title and show the plot
ax.set_title("DTM with Hillshade Overlay and Site Boundary (ChatGPTs Version)")
plt.show()
```
